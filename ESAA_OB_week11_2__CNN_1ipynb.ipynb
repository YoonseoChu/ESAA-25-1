# CH5. 합성곱 신경망 1

## 5.1 합성곱 신경망
합성곱 신경망

: 이미지 전체가 아닌 이미지의 국소적 부분을 계산함으로써 시간과 자원을 절약하여 이미지의 세밀한 부분까지 분석할 수 있는 신경만



### 5.1.1 합성곱층의 필요성
- 이미지나 영상을 처리하는 데 유용함
- 합성곱층 : 이미지를 펼쳐서 분석하면 데이터의 공간적 구조를 무시하는 것을 방지하려고 도입된 것
 - image.png

### 5.1.2 합성곱 신경망 구조
합성곱 신경망 (Convolutional Neural Network, CNN) : 음성 인식이나 이미지/영상 인식에서 주로 사용되는 신경망
- 다차원 배열 데이터를 처리하도록 구성되어 컬러 이미지 같은 다차원 배열 처리에 특화됨
- 계층 다섯 개로 구성됨
 - 입력층, 합성곱층, 풀링층, 완전연결층, 출력층
- image.png
 - 합성곱층과 풀링층 거치면서 입력 이미지의 주요 특성 벡터를 추출
 - 추출된 주요 특성 벡터들은 완전 연결층 거치면서 1차원 벡터로 변환
 - 출력층에서 활성화 함수인 소프트맥스 함수를 사용하여 최종 결과 출력

#### [입력층]
입력층(iinput layer) : 입력 이미지 데이터가 최초로 거치게 되는 계층
 - 높이, 너비, 채널의 값을 갖는 3차원 데이터
 - 채널 : 흑백이면 1, 컬러면 3


#### [합성곱층]
합성곱층(convolutional layer) : 입력 데이터에서 특성을 추출하는 역할을 수행
- 특성 추출 : 입력 이미지가 들어왔을 때 이미지에 대한 특성을 감지하기 위해 커널이나 필터를 사용
- 특성 맵 : 커널/필터이 이미지의 모든 영역을 훑으면서 특성을 추출한 결과물

#### [풀링층]
풀링층 (pooling layer)

: 합성곱층과 유사하게 특성 맵의 차원을 다운 샘플링하여 연산량을 감소시키고, 주요한 특성 벡터를 추출하여 학습을 효과적으로 할 수 있음

- 최대 풀링 (max pooling) : 대상 영역에서 최댓값을 추출
- 평균 풀링 (average pooling) : 대상 영역에서 평균을 반환

#### [출력층]
- 출력층 (output layer)에서는 소프트맥스 활성화 함수가 사용됨
- 입력받을 값을 0~1 사이의 값으로 출력
- 마지막 출력층의 소프트맥스 함수를 사용하여 이미지가 각 레이블에 속할 확률 값이 출력됨 -> 가장 높은 확률 값을 갖는 레이블이 최종 값으로 선정됨

### 5.1.3. 1D, 2D, 3D 합성곱

#### [1D 합성곱]
1D 합성곱 : 필터가 시간을 축으로 좌우로만 이동할 수 있는 합성곱
- 입력(W)과 필터(k)에 대한 출력은 W
 - image.png


#### [2D 합성곱]
- 필터가 방향 두 개로 움직이는 형태
- 입력(W, H)과 필터 (k,k)에 대해 출력은 (W,H), 출력 형태는 2D 행렬
 - image.png

#### [3D 합성곱]
- 필터가 움직이는 방향이 세 개
- 입력 (W, H, L) 에 대해 필터 (k,k,d)를 적용하면 출력으로 (W,H,L)을 갖는 형태가 3D 합성곱, 출력은 3D 형태
- d < L 을 유지하는 것이 중요
 - image.png

#### [3D 입력을 갖는 2D 합성곱]
3D 입력을 갖는 2D 합성곱
