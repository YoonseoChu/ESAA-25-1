### [DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환]

#### (1) 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

* DataFrame으로 변환 시 칼럼명을 지정
* DataFrame은 기본적으로 행과 열을 가지는 2차원 데이터

* 1차원 형태의 리스트와 넘파이 ndarray를 DataFrame으로 변환

````python
import numpy as np

# 1차원 형태의 데이터를 기반으로 DataFrame을 생성할 때 칼럼명은 한 개만 필요

col_name1 = ['col1']

list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성.
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성.
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
````

* 2차원 형태의 데이터를 기반으로 DataFrame을 생성할 때

````python
# 3개의 칼럼명이 필요
col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성 후 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)

print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
````

* 딕셔너리를 DataFrame으로 변환

### [DataFrame 데이터 삭제]

drop() 매서드
* DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
* axis=0 : 로우 방향, axis=1 : 칼럼 방향
* axis = 0 삭제 : 기존 칼럼 값을 가공하여 새로운 칼럼을 만들고 삭제
* axis = 1 삭제 : 이상치 데이터를 삭제

````python
# 'Age_0' 칼럼 삭제
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
````

````python
# 원본 데이터 확인
titanic_df.head(3)
````

* inplace = False로 설정 : 'Age_0' 칼럼이 여전히 존재 (디폴트 값)
* inplace = True로 설정 : 자신의 DataFrame의 데이터를 삭제
* 여러 개의 칼럼을 삭제 : 리스트 형태로 삭제하고자 하는 칼럼 명을 입력

````python
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print('inplace=True로 drop 후 반환한 값:', drop_result)
titanic_df.head(3)
````

* 유의할 점 : drop() 시 inplace=True로 설정하면 반환 값이 None이 됨
* titanic_df = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True) : titanic_df 객체 변수를 아예 None으로 만듦

````python
# axis = 0으로 설정해 index0,1,2 로우를 삭제

pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('###before axis drop ###')
print(titanic_df.head(3))

titanic_df.drop([0,1,2], axis=0, inplace=True)

print('### after axis 0 drop ####')
print(titanic_df.head(3))
````

* axis : DataFrame의 로우를 삭제할 때는 axis=0, 칼럼을 삭제할 때는 axis=1
* 원본 DataFrame은 유지하고 드롭된 DataFrame을 새롭게 객체 변수로 받고 싶다면 Inplace=False로 설정 (디폴트 값이 False)
* 원본 DataFrame에 드롭된 결과를 적용할 경우에는 inplace=True를 적용
```
titanic_drop_df = titanic_df.drop('Age=0', axis=1, inplace=False)
```
* 원본 DataFrmae에서 드롭된 DataFrame을 다시 원본 DataFrmae 객체 변수로 할당하면 원본 DataFrame에서 드롭된 결과를 적용할 경우와 같음 (단, 기존 원본 DataFrame 객체 변수는 메모리에서 추후 제거됨.)

[reset_index()]

* DataFrame 및 Series에 reset_index() 매서드를 수행하면 새롭게 인덱스를 연속 숫자 형으로 할당하여 기존 인덱스는 'index'라는 새로운 칼럼 명으로 추가

````python
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
````

* reset_index() : 인덱스가 연속된 int 숫자형 데이터가 아닐 경우에 다시 이를 연속 int 숫자형 데이터로 만들 때 주로 사용


````python
# Series의 'Pclass' 고유 값이 식별자 인덱스 역할을 함
print('### before reset_index ###')
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:', type(value_counts))


# 연속형 숫자형 인덱스가 고유 식별자로 더 적합
new_value_counts = value_counts.reset_index(inplace=False)

print('### after reset_index ###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:', type(new_value_counts))
# Series가 아닌 DataFrame이 반환됨
````

### [apply lambda 식으로 데이터 가공]
* apply lambda : 복잡한 데이터 가공이 필요한 경우 이용

* lambda 식 : 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들어짐

````python
def get_square(a):  # 함수명과 입력 인자를 먼저 선언
    return a**2     # 함수 내에서 입력 인자를 가공한 뒤 결괏값을 return과 같은 문법으로 반환

print('3의 제곱은:', get_square(3))
````

````python
# lambda 사용
lambda_square = lambda x : x**2   # ':'로 입력 인자와 반환될 입력 인자의 계산식을 분리 (입력인자 : 계산식)
print('3의 제곱은:', lambda_square(3))
````

````python
# map() 함수 결합 : 여러 개의 값을 입력 인자로 사용해야 할 경우
a = [1,2,3]
squares = map(lambda x : x**2,a)
list(squares)
````

apply에 lambda 식을 적용해 데이터 가공

````python
# 'Name' 칼럼의 문자열 개수를 별도의 칼럼인 'Name_len'에서 생성
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
````

* if else 절 이용

````python
# 나이가 15세 미만이면 child, 그렇지 않으면 adult로 구분
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

* if else 이용할 때, if 절의 경우 if 식보다 반환 값을 먼저 기술해야 함
* else if 는 지원하지 않음


````python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'

    return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
````
